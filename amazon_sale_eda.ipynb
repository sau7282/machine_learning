```python
# 1.What is the average rating for each product category?

import pandas as pd

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

average_rating_by_category = df.groupby('category')['rating'].mean().reset_index()
average_rating_by_category_df=pd.DataFrame(average_rating_by_category)
print(average_rating_by_category)
average_rating_by_category_df
```

```python
# 2.What are the top rating_count products by category?

import pandas as pd

df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df_clean = df.dropna(subset=['rating_count'])
top_products_by_category = df_clean.loc[df_clean.groupby('category')['rating_count'].idxmax()]
top_products_by_category = top_products_by_category[['category', 'product_name', 'rating_count']]
top_products_by_category_df=pd.DataFrame(top_products_by_category)
top_products_by_category_df
```

```python
# 3.What is the distribution of discounted prices vs. actual prices?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=True).str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=True).str.replace(',', '').astype(float)

df['discounted_price'] = df['discounted_price'].astype(int)
df['actual_price'] = df['actual_price'].astype(int)

plt.figure(figsize=(12, 6))

sns.histplot(df['discounted_price'], color='blue', label='Discounted Price', kde=True, stat='density', linewidth=0)
sns.histplot(df['actual_price'], color='red', label='Actual Price', kde=True, stat='density', linewidth=0)

plt.title('Distribution of Discounted Prices vs Actual Prices', fontsize=14)
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 4.How does the average discount percentage vary across categories?
import pandas as pd
import matplotlib.pyplot as plt

df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=True).astype(int)

avg_discount_by_category = df.groupby('category')['discount_percentage'].mean().reset_index()

avg_discount_by_category = avg_discount_by_category.sort_values(by='discount_percentage', ascending=False)


plt.figure(figsize=(12, 6))
plt.bar(avg_discount_by_category['category'], avg_discount_by_category['discount_percentage'], color='skyblue')
plt.xticks(rotation=90, ha='right')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.title('Average Discount Percentage by Category')
plt.tight_layout()
plt.show()

# Display the results
print(avg_discount_by_category)
```

```python
# 5.What are the most popular product names?

import pandas as pd

product_popularity = df['product_name'].value_counts().reset_index()
product_popularity.columns = ['product_name', 'count']
top_products = product_popularity.head(10)

print("Top 10 Most Popular Product Names:")
print(top_products)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.barh(top_products['product_name'], top_products['count'], color='lightgreen')
plt.xlabel('Count')
plt.ylabel('Product Name')
plt.title('Top 10 Most Popular Product Names')
plt.gca().invert_yaxis()  # Reverse order for better readability
plt.tight_layout()
plt.show()

# output:
# Top 10 Most Popular Product Names:
#                                         product_name  count
# 0  Fire-Boltt Ninja Call Pro Plus 1.83" Smart Wat...      5
# 1  Fire-Boltt Phoenix Smart Watch with Bluetooth ...      4
# 2  Wayona Nylon Braided USB to Lightning Fast Cha...      3
# 3  Pinnaclz Original Combo of 2 Micro USB Fast Ch...      3
# 4  Ambrane Unbreakable 60W / 3A Fast Charging 1.5...      3
# 5  Amazonbasics Nylon Braided Usb-C To Lightning ...      3
# 6  Fire-Boltt Visionary 1.78" AMOLED Bluetooth Ca...      3
# 7  boAt A400 USB Type-C to USB-A 2.0 Male Data Ca...      3
# 8  Duracell USB C To Lightning Apple Certified (M...      3
# 9  Ambrane 60W / 3A Fast Charging Output Cable wi...      3
```

```python
# 6. What are the most popular product keywords?
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

keywords = df['product_name'].str.split().explode().str.lower()

stop_words = {"and", "with", "for", "the", "of", "in", "to", "on", "by", "a", "is"}
filtered_keywords = keywords[~keywords.isin(stop_words)]

keyword_counts = Counter(filtered_keywords)

keyword_df = pd.DataFrame(keyword_counts.items(), columns=['keyword', 'count'])
keyword_df = keyword_df.sort_values(by='count', ascending=False)

top_keywords = keyword_df.head(10)
print("Top 10 Most Popular Product Keywords:")
print(top_keywords)

plt.figure(figsize=(12, 6))
plt.barh(top_keywords['keyword'], top_keywords['count'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('Keyword')
plt.title('Top 10 Most Popular Product Keywords')
plt.gca().invert_yaxis()  # Reverse the order for better readability
plt.tight_layout()
plt.show()

# output:
# Top 10 Most Popular Product Keywords:
#       keyword  count
# 194         |    456
# 39          &    432
# 3         usb    377
# 9       cable    318
# 104         -    218
# 6    charging    217
# 5        fast    209
# 92    (black)    187
# 35          c    179
# 152     smart    171
```

```python
# 7.What are the most popular product reviews?

import pandas as pd

df['review_content'] = df['review_content'].str.strip() 
df['review_content'] = df['review_content'].str.lower()  
review_counts = df['review_content'].value_counts().reset_index()
review_counts.columns = ['review_content', 'count']

top_reviews = review_counts.head(10)
print("Top 10 Most Popular Product Reviews:")
print(top_reviews)

import matplotlib.pyplot as plt

top_reviews['short_review'] = top_reviews['review_content'].str[:50] + '...'

# Plot
plt.figure(figsize=(15, 8))
plt.barh(top_reviews['short_review'], top_reviews['count'], color='lightblue')
plt.xlabel('Frequency')
plt.ylabel('Review Content')
plt.title('Top 10 Most Popular Product Reviews')
plt.tight_layout()
plt.gca().invert_yaxis()  
plt.show()

# output:
# Top 10 Most Popular Product Reviews:
#                                       review_content  count
# 0  i am not big on camera usage, personally. i wa...      8
# 1  i ordered this cable to connect my phone to an...      7
# 2  good product,long wire,charges good,nice,i bou...      7
# 3  worked on iphone 7 and didn’t work on iphone x...      7
# 4  128gb sd card is showing 134gbdon't buy this p...      6
# 5  i purchased the 6/128gb variant.to sum  it up ...      6
# 6  [update: sept 29] boat seems to have heard the...      5
# 7  i liked it, worth the money, light weight and ...      5
# 8  about the tv - wonderful----------------------...      5
# 9  good budget mfi certified lightly cable for th...      5
```

```python
# 8.What is the correlation between discounted_price and rating?
import pandas as pd

df['discounted_price'] = (
    df['discounted_price']
    .str.replace('₹', '', regex=False)  
    .str.replace(',', '', regex=False) 
    .astype(float)                      
)

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df_cleaned = df.dropna(subset=['discounted_price', 'rating'])

correlation = df_cleaned['discounted_price'].corr(df_cleaned['rating'])
print(f"Correlation between discounted_price and rating: {correlation}")

# output:

# Correlation between discounted_price and rating: 0.12113187526066276
```

```python
# 9.What are the Top 5 categories based on the highest ratings?

import pandas as pd
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
category_avg_ratings = df.groupby('category')['rating'].mean()
top_categories = category_avg_ratings.sort_values(ascending=False).head(5)

print("Top 5 Categories Based on Highest Ratings:")
print(top_categories)

# output:
# Top 5 Categories Based on Highest Ratings:
# category
# Computers&Accessories|Tablets                                               4.6
# Computers&Accessories|NetworkingDevices|NetworkAdapters|PowerLANAdapters    4.5
# Electronics|Cameras&Photography|Accessories|Film                            4.5
# Electronics|HomeAudio|MediaStreamingDevices|StreamingClients                4.5
# OfficeProducts|OfficeElectronics|Calculators|Basic                          4.5
# Name: rating, dtype: float64
```
