```python
'''
Features:

Car_id: A unique identifier for each car in the dataset, helping to track individual car entries.

Date: The date when the car sale transaction took place, formatted as YYYY-MM-DD.

Customer Name: The name of the customer who purchased the car, represented as a string.

Gender: The gender of the customer, categorized as "Male" or "Female."

Annual Income: The customer's annual income in US dollars, represented as a numeric value.

Dealer_Name: The name of the dealership selling the car, represented as a string.

Company: The manufacturer or brand name of the car, such as "Toyota," "Ford," etc.

Model: The specific model name of the car, such as "Corolla," "Civic," etc.

* Engine: The engine type of the car, such as "V6," "14," etc.

Transmission: The type of transmission in the car, either "Manual" or "Automatic."

Color: The color of the car, represented as a string (e.g., "Red," "Blue").

Price ($): The selling price of the car in US dollars.

Dealer_No: A unique identifier for each car dealer in the dataset.

Body Style: The body style of the car, such as "Sedan," "SUV," etc.

Phone: The phone number of the customer who purchased the car.

Dealer_Region: The geographical region of the car dealer, such as "North," "South," etc.
'''
```

```python
# 2.Which car brand (Company) has the highest variation in prices, and what does this tell us about the 
# pricing trends?

import pandas as pd

# Group by 'Company' and calculate the standard deviation of prices
price_variation_by_brand = df.groupby('Company')['Price ($)'].std().reset_index()

# Rename columns for clarity
price_variation_by_brand.rename(columns={'Price ($)': 'Price Std Dev ($)'}, inplace=True)

# Find the brand with the highest price variation
highest_variation_brand = price_variation_by_brand.loc[price_variation_by_brand['Price Std Dev ($)'].idxmax()]

# Display the results
print("Price Variation by Brand:")
print(price_variation_by_brand.sort_values(by='Price Std Dev ($)', ascending=False))

print("\nBrand with the Highest Price Variation:")
print(highest_variation_brand)

# output:
# Brand with the Highest Price Variation:
# Company                   Lincoln
# Price Std Dev ($)    19658.050211
# Name: 15, dtype: object
```

```python
#3.What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?

import pandas as pd
import matplotlib.pyplot as plt

# Group by 'Transmission' and calculate descriptive statistics
price_distribution_by_transmission = df.groupby('Transmission')['Price ($)'].describe()

# Calculate the IQR for each transmission type
price_distribution_by_transmission['IQR'] = price_distribution_by_transmission['75%'] - price_distribution_by_transmission['25%']

# Display the results
print("Price Distribution by Transmission:")
print(price_distribution_by_transmission[['min', '25%', '50%', '75%', 'max', 'IQR']])

# Visualize the distribution using boxplots
plt.figure(figsize=(12, 8))
df.boxplot(column='Price ($)', by='Transmission', grid=False, patch_artist=True, showmeans=True)
plt.title('Car Price Distribution by Transmission Type', fontsize=16)
plt.xlabel('Transmission Type', fontsize=14)
plt.ylabel('Price ($)', fontsize=14)
plt.suptitle('')  # Remove automatic boxplot title
plt.tight_layout()
plt.show()

# output:
Price Distribution by Transmission:
#                  min      25%      50%      75%      max      IQR
# Transmission                                                     
# Auto          1200.0  19000.0  24000.0  35500.0  75700.0  16500.0
# Manual        1700.0  17000.0  22001.0  34000.0  85800.0  17000.0
```

```python
# 11.How does car price vary with engine size across different car colors, and which colors have 
# the highest price variation?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'Engine' size and 'Color' to calculate average price
avg_price_by_engine_color = df.groupby(['Engine', 'Color'])['Price ($)'].mean().reset_index()

# Set up the plot for price distribution by engine size and color
plt.figure(figsize=(12, 8))

# Create a boxplot to visualize price distribution by engine size and color
sns.boxplot(x='Engine', y='Price ($)', hue='Color', data=df, palette="Set2")

# Title and labels
plt.title('Car Price Distribution by Engine Size and Color', fontsize=16)
plt.xlabel('Engine Size', fontsize=14)
plt.ylabel('Price ($)', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

# Calculate the price variation by color (standard deviation or IQR)
price_variation_by_color = df.groupby('Color')['Price ($)'].agg(['std', 'mean']).reset_index()

# Identify the color with the highest price variation (standard deviation)
highest_variation_color = price_variation_by_color.loc[price_variation_by_color['std'].idxmax()]

# Display the color with the highest price variation
print("Color with the Highest Price Variation:")
print(highest_variation_color)
```

```python
# 12.Is there any seasonal trend in car sales based on the date of sale?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract temporal features: year, month, and day of the week
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day_of_Week'] = df['Date'].dt.weekday

# Optional: Create a 'Season' column based on month
seasons = {
    1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 6: 'Summer',
    7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall', 11: 'Fall', 12: 'Winter'
}
df['Season'] = df['Month'].map(seasons)

# Analyze the number of sales by month
sales_by_month = df.groupby('Month').size()

# Plot the sales trend by month (Seasonal trend)
plt.figure(figsize=(10, 6))
sales_by_month.plot(kind='bar', color='skyblue')
plt.title('Car Sales Trend by Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Sales', fontsize=14)
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.tight_layout()
plt.show()

# Analyze sales by day of the week
sales_by_day_of_week = df.groupby('Day_of_Week').size()

# Plot the sales trend by day of the week
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_by_day_of_week.index, y=sales_by_day_of_week.values, palette="viridis")
plt.title('Car Sales Trend by Day of the Week', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Number of Sales', fontsize=14)
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], rotation=45)
plt.tight_layout()
plt.show()

# Analyze sales by season (if 'Season' column is created)
sales_by_season = df.groupby('Season').size()

# Plot the sales trend by season
plt.figure(figsize=(10, 6))
sales_by_season.plot(kind='bar', color='lightgreen')
plt.title('Car Sales Trend by Season', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Number of Sales', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 14.What is the correlation between car price, engine size, and annual income of customers, 
# and how do these features interact?

# Step 1: Convert Engine column to numeric if it's in string format (e.g., '1.5L', '2.0L')
df['Engine'] = df['Engine'].str.extract('(\d+\.?\d*)').astype(float)

# Step 2: Select relevant columns for analysis
df_sub = df[['Price ($)', 'Engine', 'Annual Income']]

# Step 3: Calculate the correlation matrix
correlation_matrix = df_sub.corr()

# Step 4: Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix: Car Price, Engine Size, and Annual Income', fontsize=16)
plt.tight_layout()
plt.show()

# Step 5: Visualize interactions between variables with pair plots
sns.pairplot(df_sub, kind='reg', plot_kws={'line_kws':{'color':'red'}})
plt.suptitle('Interactions between Car Price, Engine Size, and Annual Income', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#15. How does the average car price vary across different car models and engine types?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Clean the 'Engine' column (handle missing or invalid values)
df['Engine'] = df['Engine'].fillna('Unknown')  # Replace missing values with 'Unknown'
df['Engine'] = df['Engine'].astype(str)  # Ensure the 'Engine' column is treated as a string

# Step 2: Group by Car Model and Engine Type to calculate average price
average_price_by_model_engine = df.groupby(['Model', 'Engine'])['Price ($)'].mean().reset_index()

# Step 3: Visualize the results with a bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x='Model', y='Price ($)', hue='Engine', data=average_price_by_model_engine, palette='viridis')

# Title and labels
plt.title('Average Car Price by Model and Engine Type', fontsize=16)
plt.xlabel('Car Model', fontsize=14)
plt.ylabel('Average Price ($)', fontsize=14)
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
